```python
def gen(n):
    pokemon_data = pd.read_csv("pokemon.csv")
    g1 = pokemon_data[pokemon_data["#"] < 152]
    g2 = pokemon_data[(pokemon_data["#"] > 151) & (pokemon_data["#"] < 252)]
    g3 = pokemon_data[(pokemon_data["#"] > 251) & (pokemon_data["#"] < 387)]
    g4 = pokemon_data[(pokemon_data["#"] > 386) & (pokemon_data["#"] < 494)]
    g5 = pokemon_data[(pokemon_data["#"] > 493) & (pokemon_data["#"] < 650)]
    g6 = pokemon_data[(pokemon_data["#"] > 649) & (pokemon_data["#"] < 722)]
    g7 = pokemon_data[(pokemon_data["#"] > 721) & (pokemon_data["#"] < 810)]
    g8 = pokemon_data[(pokemon_data["#"] > 809) & (pokemon_data["#"] < 906)]
    g9 = pokemon_data[(pokemon_data["#"] > 905)]
    list = [g1,g2,g3,g4,g5,g6,g7,g8,g9]
    try:
        return list[n-1]
    except Exception as e:
        print("there are only 9 generation")
```
